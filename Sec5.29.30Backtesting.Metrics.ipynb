# Imports

```python
# Import necessary libraries
import polars as pl
import yfinance as yf
from datetime import datetime
import numpy as np
from pandas_datareader.data import DataReader
import matplotlib.pyplot as plt
```

# Data Extraction

```python
# Define dates and symbol
start_date = '2017-01-01'
end_date = '2022-06-01'
symbol = 'SPY'
```

```python
# Use yfinance to download the data
df_yf = yf.download(symbol, start=start_date, end=end_date, interval='1d')

# Convert to Polars DataFrame
df = pl.from_pandas(df_yf.reset_index())  # use reset_index() to ensure the date becomes a column in Polars

print(df.head(3))
print(df.tail(3))
```

# Feature Adjustments

```python
# Add moving averages
df = df.with_columns([
    pl.col('Adj Close').rolling_mean(window_size=12).alias('MA_12'),
    pl.col('Adj Close').rolling_mean(window_size=21).alias('MA_21')
])

df.head(20)
```

# Add Signal

```python
# Generate signals
df = df.with_columns(
    pl.when(pl.col('MA_12') > pl.col('MA_21')).then(1).otherwise(0).alias('Signal')
)

df = df.with_columns(
    pl.col('Signal').shift(1).alias('Shifted_Signal')
)
df.head(5)
```

### Signal Generation Explanation

The following code generates trading signals based on a moving average crossover strategy:
pl.when(pl.col('MA_12') > pl.col('MA_21')): Checks if the 12-day moving average is greater than the 21-day moving average.
.then(1).otherwise(0): Assigns 1 (buy signal) if the condition is true, 0 otherwise (no signal or sell).
.alias('Signal'): Names the resulting column 'Signal'.

df.with_columns(...): This is a Polars method that allows you to add or modify columns in the DataFrame. It returns a new DataFrame with the changes applied.
pl.col('Signal'): This selects the 'Signal' column from the DataFrame.
.shift(1): This is the key operation here. It shifts all values in the 'Signal' column down by one row.

The first row of the shifted column will contain a null value.
Each subsequent row will contain the value that was previously in the row above it.


.alias('Signal'): This names the resulting column 'Signal', effectively overwriting the original 'Signal' column.

The purpose of this operation is to implement a realistic trading scenario and avoid look-ahead bias. Here's why this is important:

In a real trading scenario, you make decisions based on past information, not future information.
The original 'Signal' column likely contained trading signals based on some calculation (like moving average crossovers) using data from the same row.
By shifting the signal down by one row, you're ensuring that the signal for each day is based on the previous day's data.
This mimics the real-world scenario where you would calculate your trading signal at the end of one day, and then apply that signal to your trading decisions on the next day.

```python
# Add Returns

# Calculate log returns for the benchmark. This is done by taking the natural logarithm of the ratio of
# the 'Adj Close' price of the current day to the previous day. This helps in understanding the returns
# if we held the asset from one day to the next.
df = df.with_columns(
    (pl.col('Adj Close') / pl.col('Adj Close').shift(1)).log().alias('log_returns_bench')
)

# Calculate the cumulative sum of the log returns for the benchmark. This cumulative sum helps us to 
# understand the total log returns over the period.
df = df.with_columns(
    pl.col('log_returns_bench').cum_sum().alias('log_prod_bench')
)

# Calculate the total return for the benchmark. This is done by taking the exponential of the cumulative
# log returns and then subtracting 1. This step normalizes the returns and gives us a more understandable
# cumulative return value.
df = df.with_columns(
    (pl.col('log_prod_bench').exp() - 1).alias('prod_bench')
)

# Display the first 3 rows to verify the added columns
print(df.head(3))
```

```python
# Add Strategy Returns

# Calculate the log returns for the strategy. This is done by taking the natural logarithm of the ratio
# of the 'Open' price of the next day to the 'Open' price of the current day and multiplying by the
# shifted signal. The shifted signal determines whether to take a position based on the strategy.
df = df.with_columns(
    ((pl.col('Open').shift(-1) / pl.col('Open')).log() * pl.col('Shifted_Signal')).alias('log_returns_strat')
)

# Calculate the cumulative sum of the log returns for the strategy. This cumulative sum helps us to
# understand the total log returns over the period based on the strategy.
df = df.with_columns(
    pl.col('log_returns_strat').cum_sum().alias('log_returns_prod_strat')
)

# Calculate the total return for the strategy. This is done by taking the exponential of the cumulative
# log returns and then subtracting 1. This step normalizes the returns and gives us a more understandable
# cumulative return value for the strategy.
df = df.with_columns(
    (pl.col('log_returns_prod_strat').exp() - 1).alias('prod_strat')
)

# Display the first 3 rows to verify the added columns
print(df.tail(3))
```

```python
# Remove NaN values
df = df.drop_nulls()
```

```python
# Review the dataframe
print(f"DataFrame length: {len(df)}")
print(df.head())
```

# Calculate metrics

```python

# Calculate total return
total_return_bench = df['prod_bench'].tail(1).item()
total_return_strat = df['prod_strat'].tail(1).item()

print(f"Total Return Benchmark: {total_return_bench:.2%}")
print(f"Total Return Strategy: {total_return_strat:.2%}")

# Calculate annualized return
years = (df['Date'].max() - df['Date'].min()).days / 365.25
ann_return_bench = (1 + total_return_bench) ** (1 / years) - 1
ann_return_strat = (1 + total_return_strat) ** (1 / years) - 1

print(f"Annualized Return Benchmark: {ann_return_bench:.2%}")
print(f"Annualized Return Strategy: {ann_return_strat:.2%}")
```

```python
# Calculate Sharpe Ratio (assuming risk-free rate of 0 for simplicity)
sharpe_ratio_bench = (ann_return_bench - 0) / (df['log_returns_bench'].std() * (252 ** 0.5))
sharpe_ratio_strat = (ann_return_strat - 0) / (df['log_returns_strat'].std() * (252 ** 0.5))

print(f"Sharpe Ratio Benchmark: {sharpe_ratio_bench:.2f}")
print(f"Sharpe Ratio Strategy: {sharpe_ratio_strat:.2f}")
```

The Sharpe ratio is a key metric in evaluating the performance of an investment strategy. It measures the excess return (or risk premium) per unit of risk in an investment asset or trading strategy.
Interpretation:

The higher the Sharpe ratio, the better the risk-adjusted performance.
It allows for comparison between strategies or investments with different levels of risk.

Benchmarks:

0.5-1.0: Acceptable to good
1.0-1.5: Very good


1.5: Excellent



A Sharpe ratio above 1.5 is considered excellent, indicating that the strategy is providing strong returns relative to its volatility. It suggests that the extra risk taken on by the investment is being well-compensated with additional returns.

Important Considerations:

The Sharpe ratio shouldn't be the only metric used to evaluate a strategy.
It assumes returns are normally distributed and that volatility is a complete measure of risk, which isn't always true in real markets.
Past performance doesn't guarantee future results.
A strategy with a lower Sharpe ratio might still be valuable if it offers diversification benefits or aligns better with specific investment goals.

```python
# Calculate Maximum Drawdown
def calculate_max_drawdown(returns):
    cumulative = (1 + returns).cum_prod()
    peak = cumulative.cum_max()
    drawdown = (cumulative - peak) / peak
    return drawdown.min()

max_drawdown_bench = calculate_max_drawdown(df['log_returns_bench'])
max_drawdown_strat = calculate_max_drawdown(df['log_returns_strat'])

print(f"Maximum Drawdown Benchmark: {max_drawdown_bench:.2%}")
print(f"Maximum Drawdown Strategy: {max_drawdown_strat:.2%}")
```

```python
# Define Sharpe Ratio Calculation Function

def calculate_sharpe(returns, trading_days=255, risk_free_rate=0.01):

    # Number of trading days
    n = trading_days

    # Square root of the number of trading days
    sqrt_n = np.sqrt(n)

    # Annualized mean return
    # Mean of daily returns multiplied by the number of trading days
    mean_return = returns.mean() * n

    # Annualized standard deviation (volatility)
    # Standard deviation of daily returns multiplied by the square root of the number of trading days
    sigma = returns.std() * sqrt_n

    # Sharpe Ratio calculation
    # (Mean return - risk-free rate) divided by the annualized standard deviation
    sharpe_ratio = (mean_return - risk_free_rate) / sigma

    # Return the Sharpe Ratio rounded to 3 decimal places
    return round(sharpe_ratio, 3)

# Example usage:
# trading_days = 252  # Typically, there are 252 trading days in a year
# returns = np.array([...])  # Array of daily returns
# risk_free_rate = 0.01  # Example risk-free rate (e.g., 1% annual)
# print(calculate_sharpe_ratio(trading_days, returns, risk_free_rate))
```

```python
# Calculate and Print Sharpe Ratios

# Convert log returns to numpy arrays
log_returns_bench = df['log_returns_bench'].to_numpy()
log_returns_strat = df['log_returns_strat'].to_numpy()

# Calculate Sharpe ratios for benchmark and strategy
bench_sharpe = calculate_sharpe(log_returns_bench)
strat_sharpe = calculate_sharpe(log_returns_strat)

# Print Sharpe ratios
print(f"Benchmark Sharpe Ratio: {bench_sharpe}")
print(f"Strategy Sharpe Ratio: {strat_sharpe}")
```
